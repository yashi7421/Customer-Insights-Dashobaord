```python
import pandas as pd

df = pd.read_csv('customer_shopping_behavior_kaggle.csv')
```

```python
df.describe(include='all')
```

```python
df['reviewrating'] = df.groupby('category')['reviewrating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['agegroup'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','agegroup']].head(10)
```

```python
frequencymapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchasefrequencydays'] = df['frequencyofpurchases'].map(frequencymapping)
```

```python
df[['purchasefrequencydays','frequencyofpurchases']].head(10)
```

```python
df[['discountapplied','promocodeused']].head(10)
```
